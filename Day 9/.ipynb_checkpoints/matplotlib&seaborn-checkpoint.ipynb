```python
students = ['A','B','C','D','E','F','G']
marks = [85,90,75,20,50,40,99]
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    students,
    marks,
    color=['red', 'pink', 'blue','yellow','black','purple','green'],  # Different colors
    width=0.4                        # Smaller width (default is 0.8)
)

plt.title('BAR GRAPH')
plt.ylim(0, max(marks) + 10)

plt.show()
```

```python
std =  ['A','B','C']
mark = [85,90,75]
plt.barh(std,mark,color = ['black','yellow','brown'])

plt.show()
```

```python
data = [10, 12, 15, 18, 20, 21, 22, 23, 25]

n, bins, patches = plt.hist(data, bins=5, edgecolor='black',linewidth=0.4)

colors = ['red', 'green', 'blue', 'orange', 'purple']

for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x and y values
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)

# Create coordinate matrices
X, Y = np.meshgrid(x, y)

# Create z values
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create figure
fig = plt.figure(figsize=(8, 6))

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels
ax.set_title("3D SURFACE")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [15, 18, 28, 35, 45]

# Create figure
plt.figure(figsize=(10, 5))

# Subplot 1
plt.subplot(1, 2, 1)
plt.plot(x, y1, marker='o', label='Sales')
plt.title("Sales Graph")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

plt.xlim(1, 5)
plt.ylim(0, 50)

plt.xticks([1, 2, 3, 4, 5])
plt.yticks([0, 10, 20, 30, 40, 50])

# Subplot 2
plt.subplot(1, 2, 2)
plt.bar(x, y2, color='black', label='Profit')
plt.title("Profit Graph")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.legend()
plt.grid(True)

plt.xlim(0, 6)
plt.ylim(0, 50)

# Adjust spacing
plt.tight_layout()

# Save figure
plt.savefig("graph.png")

# Display graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array([1, 2, 3, 4])
y = np.array([1, 1, 1, 1])
z = np.zeros(4)

dx = np.ones(4) * 0.5
dy = np.ones(4) * 0.5
dz = [5, 10, 15, 20]

ax.bar3d(x, y, z, dx, dy, dz)

ax.set_title("3D Bar Graph")

plt.show()
```
