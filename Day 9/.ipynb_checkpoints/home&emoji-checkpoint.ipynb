```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc, Wedge

fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0, 0), 0.8, color='#FFD35A')
ax.add_patch(face)

# Eyes touching the eyebrow
left_eye = Wedge(
    center=(-0.28, 0.26),  # moved up
    r=0.18,
    theta1=180,
    theta2=360,
    facecolor='white',
    edgecolor='none'
)

right_eye = Wedge(
    center=(0.28, 0.26),   # moved up
    r=0.18,
    theta1=180,
    theta2=360,
    facecolor='white',
    edgecolor='none'
)

ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Pupils
ax.add_patch(Circle((-0.28, 0.22), 0.06, color='#333333'))
ax.add_patch(Circle((0.28, 0.22), 0.06, color='#333333'))

# Eyebrow (draw AFTER eyes so it overlaps them)
ax.plot(
    [-0.55, 0.55],
    [0.26, 0.26],
    color='#333333',
    linewidth=9,
    solid_capstyle='round'
)

# Mouth
mouth = Arc(
    (0, -0.55),
    0.7,
    0.5,
    theta1=20,
    theta2=160,
    linewidth=7,
    color='#333333',
    capstyle='round'  
    
)

ax.add_patch(mouth)

ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```

```python
#Advanced home design

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon, Circle

fig, ax = plt.subplots(figsize=(10, 8))

# Sky
ax.set_facecolor("skyblue")

# Ground
ax.add_patch(Rectangle((-2, -2), 14, 3, color="limegreen"))

# House body
ax.add_patch(Rectangle((2, 1), 4, 3, color="orange"))

# Roof
roof = Polygon([[1.5, 4], [4, 6.5], [6.5, 4]],
               closed=True, color="crimson")
ax.add_patch(roof)

# Door
ax.add_patch(Rectangle((3.5, 1), 1, 1.8, color="saddlebrown"))

# Door knob
ax.add_patch(Circle((4.2, 1.9), 0.07, color="gold"))

# Windows
ax.add_patch(Rectangle((2.4, 2.2), 0.8, 0.8, color="lightblue"))
ax.add_patch(Rectangle((4.8, 2.2), 0.8, 0.8, color="lightblue"))

# Window lines
ax.plot([2.8, 2.8], [2.2, 3], color="black")
ax.plot([2.4, 3.2], [2.6, 2.6], color="black")

ax.plot([5.2, 5.2], [2.2, 3], color="black")
ax.plot([4.8, 5.6], [2.6, 2.6], color="black")









ax.set_xlim(-1, 11)
ax.set_ylim(0, 9)
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```
