```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x , y, test_size = 0.25 , random_state = 42)
```

# Evaluation Metrics

## 1. Accuracy

**Definition:** Overall kitni predictions sahi hain.

Formula:

\[
Accuracy = {TP + TN}/{TP + TN + FP + FN}
\]

---

## 2. Precision

**Definition:** Jitni Positive predictions ki gayi, unme se kitni sahi thi.

Formula:

\[
Precision = {TP}/{TP + FP}
\]

---

## 3. Recall (Sensitivity)

**Definition:** Actual Positive cases me se kitne correctly identify kiye gaye.

Formula:

\[
Recall = {TP}/{TP + FN}
\]

---

## 4. F1-Score

**Definition:** Precision aur Recall ka Harmonic Mean.

Formula:

\[
F1\ Score = 2 * {Precision *Recall}/{Precision + Recall}
\]

---

# Confusion Matrix

| Actual \ Predicted | Positive (1) | Negative (0) |
|-------------------|-------------|-------------|
| Positive (1) | TP (True Positive) | FN (False Negative) |
| Negative (0) | FP (False Positive) | TN (True Negative) |

---

# Terms Used

### True Positive (TP)
Actual Positive tha aur model ne bhi Positive predict kiya.

### True Negative (TN)
Actual Negative tha aur model ne bhi Negative predict kiya.

### False Positive (FP)
Actual Negative tha lekin model ne Positive predict kar diya.

### False Negative (FN)
Actual Positive tha lekin model ne Negative predict kar diya.

---

# Summary Table

| Metric | Formula | Purpose |
|----------|----------|----------|
| Accuracy | (TP+TN)/(TP+TN+FP+FN) | Overall Performance |
| Precision | TP/(TP+FP) | Positive Prediction Accuracy |
| Recall | TP/(TP+FN) | Positive Detection Rate |
| F1-Score | 2 × (Precision × Recall)/(Precision + Recall) | Balance Between Precision & Recall |

# Mean Squared Error (MSE)

### Formula

$$
MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2
$$

Where

- $y_i$ = Actual Value
- $\hat{y_i}$ = Predicted Value
- $n$ = Total Number of Observations

**Best Value:** 0

**Lower MSE = Better Model**

# Root Mean Squared Error (RMSE)

### Formula

$$
RMSE=\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y_i})^2}
$$

OR

$$
RMSE=\sqrt{MSE}
$$

**Best Value:** 0

**Lower RMSE = Better Model**

# Mean Absolute Error (MAE)

### Formula

$$
MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y_i}|
$$

Where

- $|y_i-\hat{y_i}|$ = Absolute Error

**Best Value:** 0

**Lower MAE = Better Model**

# R² Score

### Formula

$$
R^2=1-\frac{\sum(y_i-\hat{y_i})^2}{\sum(y_i-\bar{y})^2}
$$

Where

- $y_i$ = Actual Value
- $\hat{y_i}$ = Predicted Value
- $\bar{y}$ = Mean of Actual Values

**Best Value:** 1

**Higher R² = Better Model**

| Metric | Formula | Best Value |
|---------|---------|------------|
| MSE | $\frac{1}{n}\sum(y-\hat y)^2$ | 0 |
| RMSE | $\sqrt{MSE}$ | 0 |
| MAE | $\frac{1}{n}\sum|y-\hat y|$ | 0 |
| R² Score | $1-\frac{SS_{res}}{SS_{tot}}$ | 1 |
