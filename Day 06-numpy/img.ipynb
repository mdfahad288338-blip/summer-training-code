```python
import numpy as np
```

```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt
```

```python
img = Image.open("image1.jpg")
```

```python
img
```

```python
img1 = np.array(img)
```

```python
img1[:,:,0] = np.clip(img1[:,:,0]+200,20,20)
```

```python
h, w = img1.shape[:2]
mh, mw = h//2, w//2

img2 = img1.copy()

# Top Left -> More Red
img2[:mh, :mw, 0] = np.clip(img2[:mh, :mw, 0] + 255, 0,0)

# Top Right -> More Green
img2[:mh, mw:, 1] = np.clip(img2[:mh, mw:, 1] + 80, 0, 255)

# Bottom Left -> More Blue
img2[mh:, :mw, 2] = np.clip(img2[mh:, :mw, 2] + 80, 0, 255)

# Bottom Right -> More Yellow (Red + Green)
img2[mh:, mw:, 0] = np.clip(img2[mh:, mw:, 0] + 80, 0, 255)
img2[mh:, mw:, 1] = np.clip(img2[mh:, mw:, 1] + 80, 0, 255)

plt.imshow(img2)
plt.axis("off")
plt.show()
```
